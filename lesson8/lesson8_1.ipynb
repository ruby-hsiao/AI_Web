```python
# with as 就無需用close()
with open("names.txt", encoding="utf-8") as file:
    print(file.readline(), end='')
    print(file.readline(), end='')
file.closed
```

```python
# read & convert to list
with open("names.txt", encoding="utf-8") as file:
    content:str = file.read()
    names:list[str] = content.split(sep='\n')
    print(names)

# use names[] ->subscript
print(names[0])
print(names[-1])

# slice
print(names[:5])
print(names[-5:])


for name in names:
    print(name)
```

```python
from csv import DictReader

with open("aqx_p_488.csv", encoding='utf-8-sig', newline='\n') as aqx:
    reader:DictReader = DictReader(aqx)
    sites:list[dict] = list(reader)
    print(sites[0])

    for site in sites:
        site['aqi'] = int(site['aqi']) # 強制轉型
        print(site['sitename'], end=' ')
        print(site['county'], end=' ')
        print(type(site['aqi']), end = ' ')
        print(site['status'])
```
