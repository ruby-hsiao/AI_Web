```python
from pydantic import BaseModel, Field, field_validator

class Site1(BaseModel):
    公司名稱:str = Field(alias="Name_Ins")
    登記編號:str = Field(alias="No")
    地址:str = Field(alias="Address")
    統編:str = Field(alias="tax_ID_number")
    緯度:float = Field(alias="wgs84aX")
    經度:float = Field(alias="wgs84aY")

    @field_validator("緯度","經度", mode='before')
    @classmethod
    def chk_xy(cls, value):
        if value == '':
            print('空字串')
            return 0.0
        else:
            return value

class Factorys1(BaseModel):
    records:list[Site1]
```

```python
import json
from pprint import pprint # 從pprint module import pprint function 
with open('ntc_factorys_food.json', encoding='utf-8') as file:
    content:str = file.read() # convert to string
    # json.loads(str) # string to Python data type
    all_data = json.loads(content)

fac:Factorys1 = Factorys1.model_validate(all_data['result'])

i = 0
for rec in fac.records:
    i += 1
    print(i, rec)
    if rec.經度 == 0.0:
        print("x=0.0 stop program")
        break
```
